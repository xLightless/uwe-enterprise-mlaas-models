```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
df = pd.read_csv("Synthetic_Data.csv")
```

```python
# First 5 rows
print("--- First 5 rows ---")
print(df.head())

# DataFrame Information
print("\n--- DataFrame Information ---")
print(df.info())

# Shape
print("\n--- DataFrame Shape ---")
print(df.shape)
```

```python
# Descriptive statistics for numerical columns
print("\n--- Descriptive Statistics (Numerical Columns) ---")
print(df.describe())
```

```python
# Descriptive statistics for categorical columns
print("\n--- Descriptive Statistics (Categorical Columns) ---")
print(df.describe(include='object'))
```

```python
# Count rows with missing values
rows_with_missing = df.isnull().any(axis=1).sum()

print(f"Number of rows with missing values: {rows_with_missing}")
```

```python
# Analyze distribution of numerical features
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()

    print(f"\n--- Descriptive Statistics for {col} ---")
    print(df[col].describe())
```

```python
# Analyze distribution of categorical features
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    print(f"\n--- Value Counts for {col} ---")
    print(df[col].value_counts())
    print(f"--- Percentage Distribution for {col} ---")
    print(df[col].value_counts(normalize=True) * 100)
```

```python
# Spread analysis of numerical features
print("\n--- Spread Analysis of Numerical Features ---")
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    variance = df[col].var()
    std_dev = df[col].std()
    print(f"\n--- {col} ---")
    print(f"Variance: {variance:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")

    plt.figure(figsize=(8, 5))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col} (Visualizing Spread)')
    plt.ylabel(col)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Correlation Analysis ---
# Assumption: You have a pandas DataFrame named 'df' loaded with your data.

print("\n--- Correlation Analysis (Numerical Features) ---")

# Select only numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Check if there are at least two numerical columns to correlate
if len(numerical_cols) > 1:
    # Calculate the correlation matrix
    correlation_matrix = df[numerical_cols].corr()

    print("\n--- Correlation Matrix ---")
    print(correlation_matrix)

    # Visualize the correlation matrix using a heatmap
    plt.figure(figsize=(10, 8)) # Adjust size as needed for readability
    sns.heatmap(correlation_matrix,
                annot=True,      # Show the correlation values on the heatmap
                cmap='coolwarm', # Use a diverging color map (blue-white-red)
                fmt=".2f",       # Format annotations to two decimal places
                linewidths=.5)   # Add lines between cells
    plt.title('Correlation Matrix of Numerical Features')
    plt.show() # Display the heatmap

    # Optional: Add interpretation guidance
    print("\n--- Interpretation Guide ---")
    print("Values close to +1: Strong positive linear correlation.")
    print("Values close to -1: Strong negative linear correlation.")
    print("Values close to 0: Weak or no linear correlation.")
    print("Look for high correlations (e.g., > 0.7 or < -0.7) between features, which might indicate redundancy.")
    print("If one of these is a target variable, its correlation with others shows their linear relationship to the target.")

else:
    print("\nSkipping correlation analysis: Need at least two numerical columns in the DataFrame.")
```

```python
# Hints at feature importance
print("\n--- Hints at Feature Importance ---")

# Variance of numerical features
print("\n--- Variance of Numerical Features (Lower variance might mean less importance) ---")
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    variance = df[col].var()
    print(f"{col}: Variance = {variance:.2f}")

# Uniqueness of categorical features
print("\n--- Uniqueness of Categorical Features (Higher uniqueness might mean more specific information) ---")
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns
for col in categorical_cols:
    unique_count = df[col].nunique()
    total_count = len(df)
    uniqueness_ratio = unique_count / total_count
    print(f"{col}: Unique Count = {unique_count}, Uniqueness Ratio = {uniqueness_ratio:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nonlinearity_tests import test_nonlinearity

# Load the CSV file
df = pd.read_csv("Synthetic_Data.csv")

# Display basic info about the dataset
print("--- First 5 rows ---")
print(df.head())
print("\n--- DataFrame Shape ---")
print(df.shape)

# Get numerical columns for analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# For this example, let's assume we want to test if 'Claim Amount' is linearly related
# to other numerical variables. If your column names are different, adjust accordingly.
target_column = 'Claim Amount'  # Replace with your actual target column name

# Remove target from predictors
predictor_columns = [col for col in numerical_cols if col != target_column and 'Date' not in col]

print(f"\nTesting non-linearity with target: {target_column}")
print(f"Predictor columns: {predictor_columns}")

# Run the non-linearity tests
results = test_nonlinearity(df, predictor_columns, target_column)

# Display the overall result
print("\n=== SUMMARY ===")
print(f"Non-linearity score: {results.get('nonlinearity_score', 'N/A')}")
print(f"Interpretation: {results.get('interpretation', 'N/A')}")

# You could also test specific pairs of variables
print("\n=== TESTING INDIVIDUAL RELATIONSHIPS ===")
for predictor in predictor_columns[:3]:  # Testing first 3 predictors for brevity
    print(f"\nTesting {predictor} vs {target_column}:")
    individual_result = test_nonlinearity(df, [predictor], target_column)
    print(f"Non-linearity score: {individual_result.get('nonlinearity_score', 'N/A')}")
    print(f"Interpretation: {individual_result.get('interpretation', 'N/A')}")
```
